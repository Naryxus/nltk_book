## Exercise 19
We only observe the arguments in the ```sorted()```-methods. The expression 
```python
set(w.lower() for w in text1)```
firstly converts all capitals into lower case and then deletes all duplicates. The expression
```python
[w.lower() for w in set(text1)]```
firstly deletes all duplicates instead and then converts all capitals into lower case.
The first expression converts for example all 'The' into 'the' and then deletes all duplicates. Thus the second expression contains 'The' and 'the' before converting them into lower case. Thus usually the second expression returns a larger value. The values would be equal, if the text contains just lower case words.

````python
print('len(set(w.lower() for w in text1)) = ' + str(len(set(w.lower() for w in text1))))
print('len([w.lower() for w in set(text1)]) = ' + str(len([w.lower() for w in set(text1)])))
````

## Exercise 20
As given in the definition of table 4.2 of the first chapter,
```python
s.isupper()``` 
tests if ```s``` contains cased characters and all are uppercase. This test behaves different to the expression
```python
not s.islower()```
because of De Morgan's laws. ```s.islower()``` tests if ```s``` contains cased characters and all are lowercase. According to De Morgan's laws the negated test is true, if ```s``` doesn't contain cased characters or not all of them are lowercase. Thus the second expression would be true, if the string is empty or the string contains at least one cased character. For these examples the first expression would evaluate to ```False```.

````python
strings = ['aaa', '', 'AAA', 'AaA']
for string in strings:
    if len(string) > 0:
        print(string + '.isupper() = ' + str(string.isupper()))
        print('not ' + string + '.islower() = ' + str(not string.islower()))
    else:
        print("''.isupper() = " + str(string.isupper()))
        print("not ''.islower() = " + str(not string.islower()))
````

## Exercise 26
The expression
```python
sum(len(w) for w in text1)```
firstly creates a list of the lengths of all words in ```text1``` and then sum all lengths up, i.e. the number of all characters in ```text1```.
To calculate the average word length, the above sum has to be divided by the number of words in the text, which is ```len(text1)```.

````python
number_of_characters = sum(len(w) for w in text1)
print('Number of characters = ' + str(number_of_characters))

average_word_length = number_of_characters / len(text1)
print('Average word length = ' + str(average_word_length))
````
